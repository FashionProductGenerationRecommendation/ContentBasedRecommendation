```python
user_ids = range(1, 400)
```

```python
import pandas as pd
train = pd.read_json("dummy_data_v2.json", lines = True)
```

```python
item_ids = list(train["ItemID"])
```

```python
import random
result = []
for user in user_ids:
    item_rating_list = []
    num_items = random.randint(100, 200)
    sampling = random.sample(item_ids, k=num_items)
    for item in sampling:
        rating = random.randint(1,5)
        result.append(((user, item), rating))
```

```python
import pickle
with open('item_user_rating.json', 'wb') as fp:
    pickle.dump(result, fp)
```

```python
with open('item_user_rating.txt', mode='wt', encoding='utf-8') as myfile:
    for lines in result:
        myfile.write('\n'.join(str(line) for line in lines))
        myfile.write('\n')
```

```python
with open('item_user_rating.json', 'rb') as fp:
    r = pickle.load(fp)
```
